```python
compared_police_eth1 = data1.groupby('Officer-defined ethnicity')['Type'].count().reset_index()
compared_police_eth2 = data2.groupby('Officer-defined ethnicity')['Type'].count().reset_index()
compared_police_eth3 = data3.groupby('Officer-defined ethnicity')['Type'].count().reset_index()
compared_police_eth4 = data4.groupby('Officer-defined ethnicity')['Type'].count().reset_index()
```

```python
def swap_rows(tab, eth1='Asian', eth2='Other'):
    # Find the indices of 'Asian' and 'Other'
    asian_index = tab[tab['Officer-defined ethnicity'] == eth1].index[0]
    other_index = tab[tab['Officer-defined ethnicity'] == eth2].index[0]

    # Get the rows
    asian_row = tab.loc[asian_index].copy()
    other_row = tab.loc[other_index].copy()

    # Swap the rows
    tab.loc[asian_index] = other_row
    tab.loc[other_index] = asian_row

    return tab
```

```python
compared_police_eth1 = swap_rows(compared_police_eth1)
compared_police_eth2 = swap_rows(compared_police_eth2)
compared_police_eth3 = swap_rows(compared_police_eth3)
compared_police_eth4 = swap_rows(compared_police_eth4)
```

```python
# Calculate male and female counts for each month
male_counts = [data1[data1['Gender'] == 'Male'].shape[0],
               data2[data2['Gender'] == 'Male'].shape[0],
               data3[data3['Gender'] == 'Male'].shape[0],
               data4[data4['Gender'] == 'Male'].shape[0],
               data5[data5['Gender'] == 'Male'].shape[0]]

female_counts = [data1[data1['Gender'] == 'Female'].shape[0],
                 data2[data2['Gender'] == 'Female'].shape[0],
                 data3[data3['Gender'] == 'Female'].shape[0],
                 data4[data4['Gender'] == 'Female'].shape[0],
                 data5[data5['Gender'] == 'Female'].shape[0]]

months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

# Create angles for the radar chart
angles = [n / float(len(months)) * 2 * pi for n in range(len(months))]
angles += angles[:1]

# Create the radar chart
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 6),
                       subplot_kw=dict(polar=True))

plt.xticks(angles[:-1], months, color='grey', size=12)

# Determine the maximum value for the y-axis limit
max_val = max(max(male_counts), max(female_counts))

plt.yticks(np.linspace(0, max_val, 6)[1:], [str(int(x)) for x in np.linspace(0, max_val, 6)[1:]],
           color='grey', size=12)


plt.ylim(0, max_val)
ax.set_rlabel_position(30)

# Plot male data
male_values = male_counts + male_counts[:1]
ax.plot(angles, male_values, linewidth=1,
        linestyle='solid', label='Male')
ax.fill(angles, male_values, 'skyblue', alpha=0.4)

# Plot female data
female_values = female_counts + female_counts[:1]
ax.plot(angles, female_values, linewidth=1,
        linestyle='solid', label='Female')
ax.fill(angles, female_values, 'lightpink', alpha=0.4)

plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

plt.title('Comparison of Male and Female Cases Across Months (Jan-May 2025)')
plt.show()
```

```python
# Create a DataFrame with Month, Male, and Female columns
gender_month_counts = pd.DataFrame({
    'Month': months,
    'Male': male_counts,
    'Female': female_counts
})

display(gender_month_counts)
```

```python


# Combine data from all months and add a 'Month' column
data_combined = pd.concat([data1.assign(Month='Jan'),
                           data2.assign(Month='Feb'),
                           data3.assign(Month='Mar'),
                           data4.assign(Month='Apr'),
                           data5.assign(Month='May')])


# Group by Month and Gender, then count the cases
gender_monthly_counts = data_combined.groupby(['Month', 'Gender']).size().unstack(fill_value=0)


# Prepare data for the radar chart
categories = gender_monthly_counts.index.tolist()
N = len(categories)


# What will be the angle of each axis in the plot? (divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]


# Initialise the spider plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 6),
                       subplot_kw=dict(polar=True))


plt.xticks(angles[:-1], categories, color='grey', size=12)
ax.set_yticks([])
plt.yticks([50, 100, 150, 200, 300], ["50", "100", "150", "200", "300"], color="grey", size=10)
plt.ylim(0, 300)


# Plot each gender
for gender in gender_monthly_counts.columns:
    values = gender_monthly_counts[gender].tolist()
    values += values[:1]
    ax.plot(angles, values, linewidth=1, linestyle='solid', label=gender)
    ax.fill(angles, values, alpha=0.25)


plt.title('Number of Cases by Gender and Month', size=16, color='blue', y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()

```
